### Question 2

Ecrire une fonction <code>rendu_glouton</code> qui renvoie la liste des pièces renvoyées par l'algorithme glouton qui consiste à rendre la somme la plus grande inférieure ou égale à la somme qu'il reste à rendre, puis la tester avec l'exemple de la partie 1.
</div>

```python

def rendu_glouton(P, s):
        ''' 
        In : une liste P d'entiers triée par ordre décroissant et un entier s
        Out: liste de pièces de P dont la somme vaut s renvoyée par l'algo glouton
        '''
        rendu = []
        for piece in P:
            while s >= piece:
                s -= piece
                rendu.append(piece)
        return rendu
```

### Question 7

A l'aide de la formule précédente, écrire une fonction récursive <code>rendu_rec</code> qui prend comme arguments un ensemble $P$ d’entiers et un entier $s$ et qui renvoie le nombre minimal $n_P(s)$ de pièces de $P$ dont la somme est $s$.
</div>

```python
def rendu_rec(P,s):
    if s == 0:
        return 0
    min_pieces = float('inf')
    for p in P:
        if s - p >= 0:
            nb_pieces = rendu_rec(P, s - p)
            min_pieces = min(min_pieces, 1 + nb_pieces)
    
    return min_pieces
```

### Question 11

En reprenant le principe de mémoïsation déjà vu, écrire une fonction récursive <code>rendu_memo</code> qui prend comme arguments un ensemble $P$ d’entiers, un entier $s$ et un dictionnaire $d$ initialisé à ${0:0}$ et qui associera peu à peu des entiers $k$ vérifiant $0\le k\le s$ au nombre minimal $n_P(k)$ de pièces de $P$ dont la somme est $k$. Cette fonction renverra le nombre minimal $n_P(s)$ de pièces de $P$ dont la somme est $s$.

<i>Indication : le cas de base est $s\in d$</i>
</div>

```python
def rendu_memo(P, s, d):
    ''' in : une liste P d'entiers, un entier s, un dictionnaire d
        act : remplit le dictionnaire d de sorte que, pour des k ≤ s,
              d[k] = nombre minimal de pièces à rendre pour rendre k
        out: nombre minimal de pièces de P dont la somme est s
    '''
    # si s est deja dans le dico on la renvoie directment.
    if s in d:
        return d[s]
    min_pieces = float('inf')
    for p in P:
        if s - p >= 0:
            nb = rendu_memo(P, s - p, d)
            min_pieces = min(min_pieces, 1 + nb)
    d[s] = min_pieces
    return min_pieces
```

### Question 12

Proposer une fonction itérative <code>rendu_dyn</code> qui prend comme arguments un ensemble $P$ d’entiers et un entier $s$ et qui renvoie le nombre minimal $n_P(s)$ de pièces de $P$ dont la somme est $s$, en mettant à jour la liste $L$.
</div>

```python
def rendu_dyn(P, s):
    ''' in : une liste P d'entiers et un entier s
        out: nombre minimal de pièces de P dont la somme est s
    '''
    L = [k for k in range(s + 1)]  # L[k] pour tout k
    
    for k in range(1, s + 1):
        for p in P:
            if k - p >= 0:
                L[k] = min(L[k], 1 + L[k - p])
    return L[s]
```

## Reconstruction
Le problème du rendu de monnaie admet deux sorties possibles. On a montré comment calculer le nombre de pièces minimum, mais pas encore comment calculer la liste des pièces. On peut bien sûr adapter les algorithmes précédents pour qu’ils renvoient cette liste de pièce.
Mais il y a plus simple, en se basant sur l’algorithme de programmation dynamique. Étant donné la liste $L$ remplie à la fin de rendu_dyn, on peut déterminer cette liste de pièces. Pour cela, on part de la somme $s$, et on retient la valeur de $L[s]$. D’après la façon dont est remplie $L$, il existe une pièce $p$ telle que $L[s-p]=L[s]-1$. On cherche cette pièce $p$ et on continue jusqu’à arriver à $L[0]$

### Question 14

Modifier légérement (une seule ligne à modifier) la fonction <code>rendu_dyn</code> en une fonction <code>rendu_dyn_pour_reconstruction</code> qui prend comme arguments un ensemble $P$ d’entiers et un entier $s$ et qui renvoie la liste $L$ de taille $s+1$ telle que pour tout entier $0\le k\le s$, $L[k]$ soit le nombre minimal $n_P(k)$ de pièces de $P$ dont la somme est $k$.
</div>

```python
def rendu_dyn_pour_reconstruction(P, s):
    ''' in : une liste P d'entiers et un entier s
        out: la liste L telle que pour tout entier 0<=k<=s,
             L[k] = nombre minimal de pièces de P dont la somme est k
    '''
    L = [k for k in range(s + 1)]
    for k in range(1, s + 1):
        for p in P:
            if k - p >= 0:
                L[k] = min(L[k], 1 + L[k - p])
    return L
```

### Question 15

Ecrire une fonction <code>rendu_reconstruction</code> qui prend comme arguments un ensemble $P$ d’entiers et un entier $s$ et qui renvoie la liste de pièces $L_P(s)$ de $P$, minimale, dont la somme dont la somme est $s$.
</div>

```python
def rendu_reconstruction(P, s):
    ''' in : une liste P d'entiers et un entier s
        out: liste de pièces de P, minimale, dont la somme vaut s
    '''
    L = rendu_dyn_pour_reconstruction(P, s)
    coins_used = []
    while s > 0:
        for p in P:
            if s - p >= 0 and L[s - p] == L[s] - 1:
                coins_used.append(p)
                s -= p
                break
    return coins_used
```

Ecrire une fonction <code>sembleCanonique</code> qui prend comme arguments des entiers $p_{max}$, $k_{max}$ et $s_{max}$ et qui renvoie vrai si l'algorithme glouton renvoie une solution optimale pour tout $1\le p\le p_{max}$, tout $0\le k\le k_{max}$ et tout $0\le s\le s_{max}$, et faux sinon.
 
<i> Indication : il suffit d'effectue 3 boucles emboîtées et de renvoyer faux si l'algorithme glouton ne renvoie pas une solution optimale</i>
</div>

```python
def sembleCanonique(p_max, k_max, s_max):
    for p in range(1, p_max + 1):
        for k in range(0, k_max + 1):
            coins = [p**i for i in range(k, -1, -1)]
            for s in range(1, s_max + 1): 
                if rendu_glouton(coins, s) != rendu_dyn(coins, s):
                    return False
    return True

print(sembleCanonique(5, 5, 100))
```
